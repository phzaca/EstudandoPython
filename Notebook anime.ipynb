#### 26. Implemente a detecção de tendências: Qual gênero mais cresceu em número de animes de uma década para outra?

```python
gen_tmp = dados[['Genres','Aired']]
gen_tmp['Genres'] = gen_tmp['Genres'].str.split(',')
gen_tmp = gen_tmp.explode('Genres')
gen_tmp[['start','end']]=gen_tmp['Aired'].str.split('to',expand=True)
gen_tmp.drop(columns=['Aired','end'],inplace=True)
gen_tmp
```

```python
gen_tmp['start'] = gen_tmp['start'].str.replace(r'\s+', ' ', regex=True).str.strip()
gen_tmp['start'] = pd.to_datetime(gen_tmp['start'], errors='coerce')
gen_tmp = gen_tmp.reset_index(drop=True).sort_values('start')
gen_tmp['start'] = gen_tmp['start'].dt.year.astype('Int64')
gen_tmp.dropna(inplace=True)
gen_tmp.reset_index(drop=True,inplace=True)
gen_tmp
```

```python
decadas=[]
for i in range((min(gen_tmp['start'])//10)*10,(max(gen_tmp['start'])//10+2)*10,10):
    decadas.append(i)
decadas
```

```python
resultado = []
for i in range(len(decadas)-1):
    resultado = (gen_tmp['start'] >= decadas[i]) & (gen_tmp['start']<decadas[i+1])
    gen_tmp.loc[resultado,'decada'] = decadas[i]
gen_tmp['decada'] = gen_tmp['decada'].astype(int)
gen_tmp
```

```python
gen_tmp.sort_values('decada',ascending=True,inplace=True)
gen_tmp
```

```python
pivotado = pd.pivot_table(
    gen_tmp,
    values = 'start',
    index = 'decada',
    columns = 'Genres',
    aggfunc = 'count',
    observed = True,
    fill_value=0
)

pivotado
```

```python
generos = pivotado.columns.to_list()
decadas = pivotado.index.to_list()

crescimento = []

for i in range(1,len(decadas)):
    decada_atual = pivotado.loc[decadas[i]]
    decada_anterior = pivotado.loc[decadas[i-1]]
    
    cresc = np.where(decada_anterior!=0,
                     decada_atual/decada_anterior,
                     np.nan).round(2)
    
    cresc_series = pd.Series(cresc, index= generos, name= decadas[i])
    crescimento.append(cresc_series)

crescimento = pd.DataFrame(crescimento)
crescimento
```

```python
col_max = crescimento.max().idxmax()
idx_max = crescimento[col_max].idxmax()

crescimento.loc[[idx_max], [col_max]]
```

```python
plt.figure(figsize=(18,10))
sns.heatmap(crescimento, cmap='Blues', annot=True, linewidths=1)
plt.tight_layout()
plt.show()
```

### BRABO 

Implemente, só com Pandas, uma lógica para prever o “Type 1” de um Pokémon a partirde seus atributos numéricos (ex: HP, Attack, Defense etc). Dica: pode ser baseado namédia dos atributos por tipo.

```python
df_analise = df[['Name','Type 1']].reset_index(drop=True)
df_analise['media'] = df[['HP','Attack','Defense','Sp. Atk','Sp. Def']].mean(axis= 1)
df_analise = df_analise.sort_values(['Type 1','media'],ascending=[True,False]).reset_index(drop=True)

tipo = list(df_analise['Type 1'].unique())

plt.figure(figsize=(18,8))
for t in tipo:
    df_tipo = df_analise[df_analise['Type 1']== t]
    plt.scatter(df_tipo.index, df_tipo['media'], label= t)
plt.legend()
plt.savefig('media_tipo.png')
plt.show()
```

```python
tipo = list(df_analise['Type 1'].unique())

valores_por_tipo = {}
for t in tipo:
    medias = df_analise.loc[df_analise['Type 1'] == t, 'media'].tolist()
    valores_por_tipo[t] = medias

max_len = max(len(l) for l in valores_por_tipo.values())
for t in valores_por_tipo:
    l = valores_por_tipo[t]
    l.extend([np.nan] * (max_len - len(l)))
    valores_por_tipo[t] = l

df_novo = pd.DataFrame(valores_por_tipo)

df_novo.describe()
```

Analise se Pokémon de certas gerações têm “Total” sistematicamente mais alto oubaixo que as demais (faça gráficos por geração).

```python
data_boxplot = [total_gen[total_gen['Generation'] == g]['Total'] for g in gen]
plt.boxplot(data_boxplot, tick_labels=[f'Gen {g}' for g in gen])
plt.show()
```
